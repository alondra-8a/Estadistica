```python
from sklearn.linear_model import LinearRegression
from scipy.stats import pearsonr
import numpy as np
```

```python
from sklearn.linear_model import LinearRegression
from scipy.stats import pearsonr
import numpy as np

X1 = df1[["HorasEstudio"]]
y1 = df1[["Calificacion"]]

X2 = df2[["HorasEstudio"]]
y2 = df2[["Calificacion"]]

X3 = df3[["HorasEstudio"]]
y3 = df3[["Calificacion"]]

modelo1 = LinearRegression()
modelo2 = LinearRegression()
modelo3 = LinearRegression()

modelo1.fit(X1, y1)
modelo2.fit(X2, y2)
modelo3.fit(X3, y3)

pendiente = modelo1.coef_[0][0]
intercepto = modelo1.intercept_[0]

print(f"Ecuación Grupo: y = {pendiente:.2f}x + {intercepto:.2f}")


corr1, _ = pearsonr(df1["HorasEstudio"], df1["Calificacion"])
corr2, _ = pearsonr(df2["HorasEstudio"], df2["Calificacion"])
corr3, _ = pearsonr(df3["HorasEstudio"], df3["Calificacion"])

print(f"Coeficiente de correlación de Pearson Grupo 1: {corr1:.3f}")
print(f"Coeficiente de correlación de Pearson Grupo 2: {corr2:.3f}")
print(f"Coeficiente de correlación de Pearson Grupo 3: {corr3:.3f}")
```

```python
plt.scatter(df1["HorasEstudio"], df1["Calificacion"], label="Grupo 1")
plt.scatter(df2["HorasEstudio"], df2["Calificacion"], label="Grupo 2")
plt.scatter(df3["HorasEstudio"], df3["Calificacion"], label="Grupo 3")
plt.plot(df1["HorasEstudio"], modelo1.predict(X1),color="red", label="Linea de regresion")
plt.plot(df2["HorasEstudio"], modelo2.predict(X2),color="green", label="Linea de regresion")
plt.plot(df3["HorasEstudio"], modelo3.predict(X3),color="blue", label="Linea de regresion")
plt.title("Regresion Lineal")
plt.xlabel("Horas de Estudio")
plt.ylabel("Calificacion")
plt.legend()
plt.show()
```
