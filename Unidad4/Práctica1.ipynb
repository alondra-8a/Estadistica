#Ejercicio 1
Una revision de los pesos de 500 estudiantes del ITD, evidencio que la media es de 70 kg y la desviación estandar es de 4 kilogramos. Suponiendo que los valores se distribuyen normalmente. Hallar lo siguiente


*   Generar aleatoriamentre las edades de los 500 estudiantes con una media de 70 kg y una desviación estandar de 4 kg
*   Realizar el histograma de las edades
* Graficar la función de densidad de probabilidad (PDF)



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
media = 70
desv_std = 4
num_datos = 500

datos = np.random.normal(media, desv_std, num_datos)
datos[0:50]
```

```python
plt.figure(figsize=(6,4))
plt.hist(datos, bins = 15, alpha = 0.4, color = "b")
plt.title("Distribucion normal")
plt.xlabel("Peso en kg de estudiantes")
plt.grid(True)
plt.show()
```

```python
x = np.linspace(media - 3 * desv_std, media + 3 * desv_std, num_datos)

pdf = 1 / (desv_std * np.sqrt(2*np.pi))* np.exp(-(x-media)**2 / (2*desv_std**2))

```

```python
pdf.size, num_datos
```

```python
plt.figure(figsize=(6,4))
plt.plot(x, pdf, label = "Normal:(media{};)".format(np.round(media,2), np.round(desv_std) ))
plt.title("Funcion de densidad de probabilidad (PDF)- Distribucion normal")
plt.xlabel("Pesos en Kg. de los estudiantes")
plt.ylabel("Densidad de probabilidad")
plt.legend()
plt.grid(True)
plt.show
```

```python
plt.figure(figsize=(6,4))
plt.hist(datos, bins=15, alpha=0.4, density=True, color="b")
plt.plot(x, pdf, "k", linewidth=2, )

plt.title("Distribución normal")
plt.xlabel("Peso en kg de los estudiantes")
plt.grid (True)
plt.show()
```

#Ejercicio 2
Una revision de los pesos de 500 estudiantes del ITD, evidencio que la media es de 70kg y la desviación estandar es de 4 kilogramos. Suponiendo que los valores se distribuyen normalmente. Hallar lo siguiente:


*   Cuantos estudiantes pesan menos de 75 kg
*Cuantos estudiantes pesan menos de 62 kg
* Cuantos estudiantes pesan mas de 79 kg



```python
plt.figure(figsize=(6,4))

media = 70
desv_std = 4
total_estudiantes = 500

peso_limite = 79

prob_menos_75kg = norm.cdf(peso_limite, media, desv_std)


```

```python
estudiantes_menos_75kg = total_estudiantes * prob_menos_75kg
print("Probabilidad de que un estudiante pese menos de 75 kg", prob_menos_75kg)
print("Numero de estudiantes que pesan menos de 75 kg", estudiantes_menos_75kg)

x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, 1000)
y = norm.pdf(x , media, desv_std)

plt.plot (x, y, "b", linewidth = 2)
plt.fill_between(x, y, where=(x <= peso_limite), color="skyblue", alpha=0.5)
plt.title("Probabilidad de que un estudiante pese menos de  kg. \n $\mu=70$, $\sigma=4$, x=500")
plt.xlabel("Peso (kg)")
plt.ylabel ("Densidad de probabilidad")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6,4))

media = 70
desv_std = 4
total_estudiantes = 500

peso_limite = 62

prob_menos_62kg = norm.cdf(peso_limite, media, desv_std)
estudiantes_menos_62kg = total_estudiantes * prob_menos_62kg

print("Probabilidad de que un estudiante pese menos de 62 kg:",prob_menos_62kg)
print("Numero de estudiantes que pesan menos de 7r kg:", estudiantes_menos_62kg)
```

```python
x = np.linspace(media -4 * desv_std, media + 4 * desv_std, total_estudiantes)
y = norm.pdf(x, media, desv_std)

plt.plot(x, y, "b", linewidth=2)
plt.fill_between(x, y, where=(x <= peso_limite), color = "darkblue", alpha=0.5)
plt.title("Probailidad de que un estudiante pese menos de 62 kg,\n $\mu=70$, $\sigma=4$, x=500")
plt.xlabel("Peso (kg)")
plt.ylabel("Densidad de probabilidad")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6,4))

media = 70
desv_std = 4
peso_limite = 79
total_estudiantes = 500

prob_menos_79kg = norm.cdf(peso_limite, media, desv_std)
prob_mas_79kg = 1 - prob_menos_79kg
estudiantes_mas_79kg = total_estudiantes * prob_mas_79kg

print("Probabilodad de que un estudiante pese mas de 79 kg:", prob_mas_79kg)
print("Numero aproximado de estudiantes que pesan mas de 79 kg:", estudiantes_mas_79kg)
```

```python
x = np.linspace(media - 4 * desv_std, media + 4 * desv_std, total_estudiantes)
y = norm.pdf(x, media, desv_std)

plt.plot(x, y, "b", linewidth=2)
plt.fill_between(x, y, where=(x >= peso_limite), color = "darkblue", alpha=0.5)
plt.title("Probabilidad de que un estudiante pese mas de 79 kg. \n $\mu=70$, $\sigma=4$, x=500")
plt.xlabel("´Peso (Kg)")
plt.ylabel("Densidad de probabilidad")
plt.grid(True)
plt.show ()
```
